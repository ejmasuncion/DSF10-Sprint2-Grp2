```python
# ARTIST 1
artist1 = tracks_df[tracks_df['artist_name'] == 'TWICE'].copy()
# ARTIST 2
artist2 = tracks_df[tracks_df['artist_name'] == 'Red Velvet'].copy()
```

```python
feature_cols = tracks_df.columns[7:]
```

```python
# complete code to filter table for artist 1
for col in feature_cols:
    sns.histplot(artist1[col], label = 'TWICE', color = 'pink')
    sns.histplot(artist2[col], label = 'Red Velvet', color = 'red')
    
    plt.legend()             
    plt.title(col)
    plt.show()
```

Insights on audio features histograms

```python
artist1_streams = streams_df[streams_df['artist'] == 'TWICE'].loc['2018-01-01':'2018-12-31']\
    .groupby(['artist'])['streams']\
    .resample('M').sum().reset_index()\
    .sort_values('streams', ascending=False)
#df_ed = df_ed.set_index('date')
# df_ed['album_name'] = df_ed['album_id'].apply(lambda x:
#                                               'No.6' if (x == '3oIFxDIo2fwuk4lwCmFZCx') else
#                                               'Divide' if (x == '3T4tUhGYeRNVUGevb0wThu') else 'others')

artist1_streams = artist1_streams.set_index('date')
artist1_streams.head()
```

```python
artist2_streams = streams_df[streams_df['artist'] == 'Red Velvet'].loc['2018-01-01':'2018-12-31']\
    .groupby(['artist'])['streams']\
    .resample('M').sum().reset_index()\
    .sort_values('streams', ascending=False)
#df_ed = df_ed.set_index('date')
# df_ed['album_name'] = df_ed['album_id'].apply(lambda x:
#                                               'No.6' if (x == '3oIFxDIo2fwuk4lwCmFZCx') else
#                                               'Divide' if (x == '3T4tUhGYeRNVUGevb0wThu') else 'others')

artist2_streams = artist2_streams.set_index('date')
artist2_streams.head()
```

```python
df_concat = pd.concat([artist1_streams, artist2_streams])
df_concat
```

```python
df_concat.pivot(columns='artist',values='streams')#/1000000
```

```python
plt.figure(figsize=(8, 6))
ax = plt.subplot(111)

color_list = ['Red', 'Pink']
# reshape
data = df_concat.pivot(columns='artist',values='streams')#/1000000
data.plot.area(ax=ax, lw=0, color=color_list)

#plt.yticks(np.arange(0,25,5),[str(x)+'M' if x>0 else str(x) for x in np.arange(0,25,5)])
plt.ylabel('streams')

plt.xlabel('')
plt.legend(frameon=False)
```

Insights on stream of two artists
- TWICE has higher stream count
- TWICE was on the chart longer than Red Velvet

```python
features = ["danceability", "energy", "valence", "tempo", "loudness"]

columns_to_view = ['artist_name', 'track_name'] + features
```

```python
df_features_artist1 = artist1[columns_to_view].copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_features_artist1['loudness'] = scaler.fit_transform(df_features_artist1[['loudness']])
df_features_artist1['loudness'].describe()
```

```python
max_tempo = df_features_artist1['tempo'].max()
# max_loudness = df_features_artist1['loudness'].max()

df_features_artist1['tempo']= df_features_artist1['tempo']/max_tempo
# df_features_artist1['loudness']= df_features_artist1['loudness']/max_loudness

df_features1 = df_features_artist1.set_index(['track_name', 'artist_name'])

df_features_stacked1 = pd.DataFrame({'value': df_features1.stack()})

df_features_stacked1 = df_features_stacked1.reset_index()
df_features_stacked1 = df_features_stacked1.rename(columns={'level_2': 'feature'})

df_features_stacked1
```

```python
df_features_artist2 = artist2[columns_to_view].copy()
```

```python
df_features_artist2['loudness'] = scaler.fit_transform(df_features_artist2[['loudness']])
df_features_artist2['loudness'].describe()
```

```python


max_tempo = df_features_artist2['tempo'].max()
# max_loudness = df_features_artist2['loudness'].max()

df_features_artist2['tempo']= df_features_artist2['tempo']/max_tempo
# df_features_artist2['loudness']= df_features_artist2['loudness']/max_loudness

df_features2 = df_features_artist2.set_index(['track_name', 'artist_name'])

df_features_stacked2 = pd.DataFrame({'value': df_features2.stack()})

df_features_stacked2 = df_features_stacked2.reset_index()
df_features_stacked2 = df_features_stacked2.rename(columns={'level_2': 'feature'})

df_features_stacked2
```

```python
concat_features = pd.concat([df_features_stacked1, df_features_stacked2])
concat_features
```

```python
plt.figure(figsize=(8, 6))
ax = plt.subplot(111)

sns.boxplot(data=concat_features, x='feature', y='value',  hue='artist_name', ax=ax,
            hue_order=['TWICE', 'Red Velvet'], palette=['pink', 'red'])

ax.legend(loc='upper center', bbox_to_anchor=(
   0.5, -0.1), frameon=False, ncol=3)
```

Insights on stream of two artists
- Both artist has almost the same features
- TWICE has lower tempo which is surprising for their songs

```python
df_bb = streams_df[streams_df['artist'] == "TWICE"].groupby('track_name')[['streams']]\
    .resample('M').sum()
df_bb = df_bb.reset_index()
df_bb = df_bb[df_bb['date'] < '2020-01-01']
# # clean long titles
# df_bb['track_name'] = df_bb['track_name'].apply(lambda x: x.split('(')[0])\
#     .apply(lambda x: x.split(' - ')[0])
df_bb
```

```python
arr_df = df_bb.pivot(index='track_name', columns='date', values='streams')
# divide by 1M to show streams in millions
arr_df = arr_df/1000000
arr_df.fillna(0, inplace=True)
arr_df['total_streams'] = arr_df.sum(axis=1)
#arr_df = arr_df.sort_values('total_streams',ascending=False)
arr_df
```

```python
plt.figure(figsize=(8, 6))
ax = plt.subplot(111)

# get all month columns and specify format for xticks
moncols = arr_df.columns[:-1]
yymm_cols = pd.Series(moncols.values).apply(lambda x: x.strftime('%Y-%m'))

sns.heatmap(arr_df[moncols], ax=ax,
            cmap='viridis',
            cbar_kws={'label': 'million streams', 'ticks': np.arange(0, 8, 1)},
            xticklabels=yymm_cols, yticklabels=True)

plt.ylabel('')
plt.xlabel('')
```
