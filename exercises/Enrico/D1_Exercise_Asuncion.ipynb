```python
import pandas as pd
```

#### Data Prep

```python
# read and process the charts dataset
charts_df = pd.read_csv('../spotify_daily_charts.csv')
# transform date column into a datetime column
charts_df['date'] = pd.to_datetime(charts_df['date'])
charts_df.head()
```

```python
# read and process the tracks dataset
tracks_df = pd.read_csv('../spotify_daily_charts_tracks.csv')
tracks_df.head()
```

```python
# merge the charts and tracks dataset
df = charts_df.merge(tracks_df, on='track_id', how='left')

df = df.drop(columns='track_name_y')
df = df.rename(columns={'track_name_x': 'track_name'})
df.head()
```

1. What are the top 10 longest songs in the dataset

```python
# remove duplicate songs due to days
group_track_id = df.drop_duplicates(['track_name'])
# sort values
duration_ranking = group_track_id.sort_values(by = ['duration'], ascending=False)
# get top 10
duration_ranking[['track_name','artist', 'duration']].head(10)
```

2. How many hit songs has Bruno Mars had since 2017?

```python
# Create year column
df['Year'] = df['date'].dt.year
# create 2017 only df
chart_2017 = df[df['Year'] == 2017]
# create Bruno Mars df for 2017
bruno_2017 = chart_2017[chart_2017['artist_name'] == 'Bruno Mars']
# return unique track songs
hit_songs_bruno_mars = bruno_2017.track_id.unique()
# print length of df
len(hit_songs_bruno_mars)
```
